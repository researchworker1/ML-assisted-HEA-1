```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import shap
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score  
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
import math
```

```python
df = pd.read_csv('HVfeature.csv')
df.head()
```

```python
y = df['HV'].values
X =  df.drop(["HV"], axis=1)
X.head()
```

```python
model1 = RandomForestRegressor(n_estimators=300, random_state=42, max_features='log2')
model1.fit(X, y)
```

```python
scores = cross_val_score(model1, X, y, cv=5) 
print(scores.mean())
```

```python
y_pred = cross_val_predict(model1, X, y, cv=5) 
```

```python
mae = mean_absolute_error(y, y_pred)
mae
```

```python
mse = mean_squared_error(y, y_pred)
rmse = math.sqrt(mse)
rmse
```

```python
model2 = LinearRegression()
model2.fit(X,y)
```

```python
scores = cross_val_score(model2, X, y, cv=5) 
print(scores.mean())
```

```python
y_pred = cross_val_predict(model2, X, y, cv=5) 
```

```python
mae = mean_absolute_error(y, y_pred)
mae
```

```python
mse = mean_squared_error(y, y_pred)
rmse = math.sqrt(mse)
rmse
```

```python
model3 = SVR(C=1.5,kernel='linear')
model3.fit(X,y)
```

```python
scores = cross_val_score(model3, X, y, cv=5) 
print(scores.mean())
```

```python
y_pred = cross_val_predict(model3, X, y, cv=5) 
```

```python
mae = mean_absolute_error(y, y_pred)
mae
```

```python
mse = mean_squared_error(y, y_pred)
rmse = math.sqrt(mse)
rmse
```

```python
model4 = GradientBoostingRegressor(n_estimators=500, learning_rate=0.2, random_state=42)
model4.fit(X,y)
```

```python
scores = cross_val_score(model4, X, y, cv=5) 
print(scores.mean())
```

```python
y_pred = cross_val_predict(model4, X, y, cv=5) 
```

```python
mae = mean_absolute_error(y, y_pred)
mae
```

```python
mse = mean_squared_error(y, y_pred)
rmse = math.sqrt(mse)
rmse
```
